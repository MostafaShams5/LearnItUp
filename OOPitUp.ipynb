In OOP we objectify code; we try to mimic our evolved human perspective of the world the same as in code.

To make an object we should know of what we need to make an object, so here we make a `Class` that will make objects out of it. This Class has the definition of every Object, or what every Object must have by definition.

It has **Attributes** (things it is) and **Methods** (things it does):

### Attributes:
*   **Class (Static) Attribute:** All the objects have the same Attribute; it's in their core and has only one place in RAM.
*   **Instance (Object) Attribute:** All the objects have their own.

### Methods:
*   **Instance Method:** It's a Method that changes and interacts with this specific Object, not anything else.
*   **Class Method:** Interacts with the definitions that the Class has. It will affect the making of new objects; it changes the shared data (Class Attributes) of the blueprint that all objects build upon.
*   **Static Method:** It's more like a helper that no one touches. It's a standalone method that the Class Method/Instance Method might use for their functionality, and the developer uses as well to know an info (it mostly does logic or math steps).

````python
class JobApplication:
    platform = "linkedin"  # Class Attribute for all objects
    
    def __init__(self, cv_pdf, role, is_remote, summer_training_completed=False):
        self.cv_pdf = cv_pdf
        self.role = role
        self.summer_training_completed = summer_training_completed
        self.job_status = "On Site"
        
        if is_remote:
            self.job_status = "Remote"
            
    def display_summary(self):
        if self.cv_pdf:
            print(f"You're looking for a {self.job_status} role as a {self.role}.")
            print(f"Summer Training Completed: {self.summer_training_completed}")
  # @ is called a Decorator 
  # python behave differently with Decorators
    @classmethod            
    def update_platform(cls, new_platform):
        cls.platform = new_platform

    #there is no 'self' we call it after making an object that takes the argument that takes 
    @staticmethod
    def cv_score(cv_pdf):
        # Pure logic helper
        if len(cv_pdf) > 5:
            return True
        return False
    # here an instance method that uses the "@staticmethod"    
    def application_status(self):
        # Using 'self' to access both the instance data and the static tool
        return "Accepted" if self.cv_score(self.cv_pdf) else "Rejected"

# --- Testing the code ---
job1 = JobApplication("thisshouldbeenough", "ML engineer", False, summer_training_completed=True)
print 
print(JobApplication.platform)
job1.display_summary()
print(f"Status: {job1.application_status()}")
````

## Encapsulation
Encapsulation means encapsulating something because not everyone should see what is happening inside it. It hides details for security and control. With encapsulation, you don’t allow the object to access some attributes directly, though we might change these attributes through public methods.

To make an attribute private, we put `__` before its name. Then Python will throw an `AttributeError` if we create an object and try to read or overwrite it directly.

Python gives us `@property` (getters and setters) for these private attributes, so we can allow getting and setting them with our own rules and validations. Even though it is a method inside the class, it looks like an attribute when used.

````python
class Laptop:
    user_number = 0  # class variable (shared between all laptops)

    def __init__(self, brand, User_Name):
        self.brand = brand 
        self._User_Name = User_Name  # '_' means protected means PLZ don't touch but you can ofcourse (but don't fr)
        self.__Passsword = 102222  # '__' means private (python will hide it using name mangling)
        Laptop.user_number += 1  # count how many laptops created (shared counter for all objects)


    @property
    def password(self):
        # we said we don't really need getter here
        # but python forces us to create one if we want setter
        # so let's block reading the password
        raise AttributeError("Password is write-only bro don't try to read it 👀")


    @staticmethod
    def ValidUser(SMScode, userName, real_user):
        # static method = doesn't need self or have any self. inside it bucause its standalone method
        if 3452 < SMScode < 3513 and userName == real_user:
            return True
        else:
            return False


    @password.setter  # must take one argument only (the new value)
    def password(self, NewPassword):
        # setter cannot exist alone
        # it must be attached to @property
        sms = int(input('write SMS: '))
        user = input('write userName: ')

        if self.ValidUser(sms, user, self._User_Name):
            self.__Passsword = NewPassword
            print("done password changed successfully")
        else:
            print("liar verification failed")


# create objects
l1 = Laptop("HP", "Ali")
l2 = Laptop("Dell", "Sara")

print("Total laptops created:", Laptop.user_number)
print("Brand of l1:", l1.brand)
print("Brand of l2:", l2.brand)

print("\nTrying to change password for l1")
l1.password = 9999   # will ask for SMS and username

print("\nTrying to read password (should fail)")
try:
    print(l1.password)
except Exception as e:
    print("Error:", e)

print("\nAccessing protected variable (not recommended but possible)")
print(l1._User_Name)  # works but don't do this in real projects

print("\nTrying to hack private variable directly")
try:
    print(l1.__Passsword)  # this will fail
except Exception as e:
    print("Error:", e)

print("\nBut python secretly renamed it to:")
print(l1._Laptop__Passsword)  # name mangling trick
````

## Inheritance
Inheritance is when you make a class and then you make classes that get all the attributes and methods of the Parent.

The Inheritance is done by passing the parent inside the child class: `class Child(Parent)`.
That is enough to make it access the class attributes and all methods (even `__init__` of the parent if you didn't override it. If you override it, it will not know what is inside the init of the parent, and all the methods based on `__init__` attributes won't work).

So Python made `super().__init__(attributes of parent init)`. This line runs the `__init__` of the parent alongside with the child's `__init__`.

There are some names you should know:
*   **Hierarchical Inheritance (The Siblings):** This happens when ONE Parent has MULTIPLE Children.
*   **Multilevel Inheritance (The Generational Line):** This is when a Child class becomes a Parent to a new Child class. Think of it as Grandparent ➔ Parent ➔ Child.
*   **Multiple Inheritance (The Python Special):** This is the most powerful (and most dangerous—causes "The Diamond Problem") type. This is when ONE Child inherits from TWO completely different Parents. Here a problem happens when the two parents have the same method name and the child calls it; we don't know which one they want. Python solves this using **MRO (Method Resolution Order)**. Python reads it from Left to Right.

````python
class University:
    def __init__(self, year, name):
        self.year = year
        self.name = name
        
    def get_years_running(self):
        return 2026 - self.year 

class College(University):
    def __init__(self, college_name, name, year):
        self.college_name = college_name
        
        # Pushes 'year' and 'name' up to the Grandparent (University) and not just on attribute even if we just will use one
        super().__init__(year, name)
        
    def full_college_name(self):
        return f"{self.college_name} of {self.name} Uni"

# MULTILEVEL INHERITANCE: The Generational Line
# Department inherits from College, which already inherited from University.
class Department(College):
    
    def __init__(self, dept_name, college_name, name, year):
        # 1. We grab the specific attribute for this Child class
        self.dept_name = dept_name
        
        # 2. THE CHAIN REACTION: 
        # We use super() to call the Parent's (College) __init__.
        # We MUST pass college_name, name, and year because College's __init__ requires them.
        # College will then take 'name' and 'year' and pass them up to University.
        super().__init__(college_name, name, year)

    def get_full_info(self):
        # We can access self.name (Grandparent) and self.college_name (Parent) 
        # because super() successfully chained them all together.
        return f"Department: {self.dept_name} | {self.college_name} | {self.name} Uni"

# We instantiate the final Child class.
# Notice how we pass all the data required for the whole family tree right here.
cs_dept = Department('Computer Science', 'FCI', 'Menoufia', 1999)

print(cs_dept.get_full_info())
# Output: Department: Computer Science | FCI | Menoufia Uni

# Proving the Child has access to the Grandparent's methods
print(f"Years Running: {cs_dept.get_years_running()}")
# Output: Years Running: 27
````

## Polymorphism
Polymorphism is using the same method name or operator to do different jobs. It behaves differently depending on the object it is called on. It is even built-in to Python! For example, if we use the `+` operator, it will behave differently whether we pass it numbers or strings.

1.  **Duck Typing (The Python Special):** *"If it walks like a duck and quacks like a duck, it must be a duck."*
    It doesn't matter if two objects are completely unrelated. If they both have a method with the exact same name, you can treat them the same way. For example, you can call `object.run()`, and this object might be a Man or a Duck.
2.  **Polymorphism through Inheritance (Method Overriding):** This happens when we redefine a method in a child class that inherits from a parent class. While overriding does replace the parent's version for that child object, we can use `super()` to make the parent's method "ride with it" so we don't totally lose the original behavior. If you think about it, this happens a lot!
3.  **Operator Overloading (Magic/Dunder Methods):** *"Dunder = Magic = `__bla__`"*
    We can overload operators by defining special methods, like `def __add__(self, other)`. This allows us to use standard operators on custom objects that normally wouldn't support them. For example, we could do `dataset + dataset`—we can actually write the magic method to make this work!

````python
# Polymorphism: Same method/operator doing different jobs depending on the object. 
# Built right into Python (like '+' for numbers vs. strings).

class media:
    def __init__(self, player_name, supported_format):
        self.name = player_name
        self.format = supported_format
        
    def process(self):
        return f"Player: {self.name} ({self.format})"

class video(media):
    def __init__(self, video_name, duration, player_name, format_required):
        self.video_name = video_name
        self.duration = duration
        # super() just runs the parent's setup so we don't have to repeat ourselves
        super().__init__(player_name, format_required)
        
    # 2. Method Overriding: We use super() to make the parent's method "ride with it".
    def process(self):
        return f"{super().process()} | Video: {self.video_name}, {self.duration} mins"

    # 3. Operator Overloading (Dunder = Magic = __bla__): 
    # Teaches the standard '+' operator how to add custom video objects.
    def __add__(self, video2):
        return f"Combined videos: {self.video_name} + {video2.video_name} ({self.duration + video2.duration} mins)"


class music(media):
    def __init__(self, music_name, duration, player_name, format_required):
        self.music_name = music_name
        self.duration = duration
        super().__init__(player_name, format_required)
        
    # 2. Method Overriding: No super() here, so we overwrite the parent 100%.
    def process(self):
        return f"Song: {self.music_name}, {self.duration} mins"
        
    def __add__(self, song2): 
        return f"Playlist is: {self.music_name} and {song2.music_name} ({self.duration + song2.duration} mins)"


MoneyTrees = music("MoneyTrees", 3.45, "Spotify", "mp3")
MardyBum = music("MardyBum", 3.20, "Spotify", "mp3")
TravellingTips = video("TravellingTips", 39.45, "VLC", "mp4")

# (Dunder/Magic Method, __bla__())
compined_songs = MoneyTrees + MardyBum
print(compined_songs) 
print("\n")

# 1. Duck Typing: "If it walks like a duck..."
# Python doesn't care if it's music or a video. They both have a .process() method, so they both work!
print(MardyBum.process())
print(TravellingTips.process())
````

## Abstraction
Abstraction is a way of forcing a MUST to use certain necessary methods. In Abstraction, we abstract only the NECESSARY stuff. It looks like Encapsulation, but Encapsulation hides stuff for Security and Control by making some things private and others public. In Abstraction, we abstract only the NECESSARY stuff from the UNNECESSARY stuff.

In Abstraction, we define the very base class that has the MUST methods, like 'MUST make a safety check before launching a rocket.'

```python
from abc import ABC, abstractmethod

class bla(ABC): 
    pass
```

````python
from abc import ABC, abstractmethod

class Rocket(ABC): # Abstract Base Class
    # if we made abstractmethod here we MUST define __init__ in the child
    def  __init__(self, volt):
        self.volt = volt
    # this one without abstractmethod it's not a MUST to use in the Child Class
    def Sound(self):
        return "Voooooooo"
        
    @abstractmethod # here we saying any child MUST make a method called safety_check()
    def safety_check(self):
        pass


class TeslaRocket(Rocket):
    def __init__(self, volt):
        super().__init__(volt)
        
    def safety_check(self):
        if self.volt > 150 and self.volt < 500:
            return True
        return False

class NasaRocket(Rocket):
    def __init__(self, volt):
        super().__init__(volt)

Tesla = TeslaRocket(200)
# Nasa = NasaRocket(0), TypeError: Can't instantiate abstract class NasaRocket without an implementation for abstract method 'safety_check'
print (Tesla.safety_check())
print (Tesla.Sound()) # here sound worked without writing it on Tesla Class
````

*   **O - Open/Closed Principle (OCP):**
    Our code should be open for extension and closed for modification when we need to add a new feature. We do this through abstraction and inheritance; instead of modifying existing code, we create a new class based 

````python
from abc import ABC, abstractmethod

# 1. The Base Concept (Closed for modification)
class DiscountStrategy(ABC):
    @abstractmethod # A must
    def calculate(self, amount):
        pass

# 2. Extension A
class StandardDiscount(DiscountStrategy): # Inhertince here
    def calculate(self, amount):
        return amount

# 3. Extension B
class VIPDiscount(DiscountStrategy):
    def calculate(self, amount):
        return amount * 0.8

# Now, our calculator simply accepts a strategy and uses it!
class DiscountCalculator:
    def process_payment(self, discount_strategy: DiscountStrategy, amount):
        return discount_strategy.calculate(amount)




# We just extend the system with a new class if we want to add a new feature. 
class PremiumDiscount(DiscountStrategy):
    def calculate(self, amount):
        return amount * 0.91
````

*   **L - Liskov Substitution Principle (LSP):**
    *"If it looks like a duck, quacks like a duck, but needs batteries... you probably have the wrong abstraction!"*
    The idea here is: don’t create a base class and then make child classes that behave differently in the methods defined by the base class. Every method in the base class must be in the child by definition.

````python
class Document:
    def __init__(self, data):
        self.data = data

    def read(self):
        return self.data
    # not all documents can save it shouldn't be here in the first place
    def save(self, new_data):
        print("Saving data to database...")
        self.data = new_data

# We inherit from Document to get the read() logic
class ReadOnlyPDF(Document):
    def save(self, new_data):
        #  Violates LSP! We are breaking the parent's contract.
        raise Exception("Error: Cannot save a read-only PDF!")

# here is the right code:
# 1. The Base Concept (Only UNIVERSAL behaviors)
class Document:
    def __init__(self, data):
        self.data = data

    def read(self):
        return self.data

# 2. Specialized capability for writing
class WritableDocument(Document):
    def save(self, new_data):
        print("Saving data to database...")
        self.data = new_data

# 3. Read-only implementation
class ReadOnlyPDF(Document):
    pass # It just inherits read(), and doesn't pretend it can save.
````

*   **I - Interface Segregation Principle (ISP):**
    Don’t make a base/abstract class very large; a child class might not need all of its methods.

````python
from abc import ABC, abstractmethod

class CloudProvider(ABC):
    @abstractmethod
    def create_server(self): pass
    
    @abstractmethod
    def get_database(self): pass
    
    @abstractmethod
    def store_file(self): pass

# AWS can do everything, so it works fine here.
class AmazonWebServices(CloudProvider):
    def create_server(self): print("Spinning up EC2...")
    def get_database(self): print("Connecting to RDS...")
    def store_file(self): print("Saving to S3...")

#  Violates ISP! Dropbox only does storage, but is forced to implement everything.
class Dropbox(CloudProvider):
    def create_server(self): 
        raise NotImplementedError("Dropbox doesn't do servers!")
    
    def get_database(self): 
        raise NotImplementedError("Dropbox doesn't have databases!")
    
    def store_file(self): 
        print("Saving file to Dropbox...")


# now look how PROs handle it
from abc import ABC, abstractmethod

# 1. Segregated Interfaces (Small and specific)
class ServerProvider(ABC):
    @abstractmethod
    def create_server(self): pass

class DatabaseProvider(ABC):
    @abstractmethod
    def get_database(self): pass

class StorageProvider(ABC):
    @abstractmethod
    def store_file(self): pass

# 2. AWS implements all the ones it needs
class AmazonWebServices(ServerProvider, DatabaseProvider, StorageProvider):
    def create_server(self): print("Spinning up EC2...")
    def get_database(self): print("Connecting to RDS...")
    def store_file(self): print("Saving to S3...")

# 3. Dropbox only implements exactly what it uses
class Dropbox(StorageProvider):
    def store_file(self): print("Saving file to Dropbox...")
````

*   **D - Dependency Inversion Principle (DIP):**
    This one is a little hard. When writing a high-level module, it should rely on an abstract one, not a low-level one (the one relying on the abstraction).

````python
class FileLogger:
    def log(self, message):
        print(f"Writing to text file: {message}") # this is specfic

class UserService:
    def __init__(self):
        # Violates DIP! We are hardcoding the specific tool here.
        self.logger = FileLogger() 

    def create_user(self, username):
        # Core business logic...
        self.logger.log(f"Created user: {username}")


# We fix this by creating an abstraction (a Logger interface). The UserService will only care about the interface, not how the logging actually happens.
from abc import ABC, abstractmethod

# 1. The Abstraction (Blueprint)
class Logger(ABC):
    @abstractmethod
    def log(self, message):
        pass

# 2. Low-level module A (Depends on the blueprint)
class FileLogger(Logger):
    def log(self, message):
        print(f"Writing to text file: {message}")

# 3. Low-level means does the nitty-gritty details
class DatadogLogger(Logger):
    def log(self, message):
        print(f"Sending to Datadog API: {message}")

# 4. High-level module (Depends on the blueprint, not the details): High-level means its done a real thing closest to the real world stuff 
class UserService:
    # We "inject" the dependency through the constructor
    def __init__(self, logger: Logger): 
        self.logger = logger

    def create_user(self, username):
        # Core business logic...
        self.logger.log(f"Created user: {username}")



# Old way
# service = UserService(FileLogger())

# New way - swapped out effortlessly!
service = UserService(DatadogLogger()) 
service.create_user("alice_smith")
````
